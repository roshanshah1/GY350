### save cleaned CSVs to another folder

```python
import os
import pandas as pd

# Define directories
source_dir = r"H:\GY350\File Downloads"
destination_dir = r"H:\GY350\CSVs Cleaned"
os.makedirs(destination_dir, exist_ok=True)  # Ensure the directory exists

# Iterate over years from 2010 to 2023
for year in range(2010, 2024):
    source_file = os.path.join(source_dir, f"DP03_06_{year}.csv")
    destination_file = os.path.join(destination_dir, f"DP03_06_{year}.csv")
    
    if not os.path.exists(source_file):
        print(f"File not found: {source_file}")
        continue
    
    # Load and process the CSV
    cleaned_df = pd.read_csv(source_file, header=None)
    
    # Identify columns to keep (columns where row 2 has a value)
    columns_to_keep = cleaned_df.iloc[1].notna()
    
    # Drop columns that do not meet the condition
    cleaned_df = cleaned_df.loc[:, columns_to_keep]
    
    # Drop the first row (original row 1)
    cleaned_df = cleaned_df.iloc[1:].reset_index(drop=True)
    
    # Save cleaned CSV
    cleaned_df.to_csv(destination_file, index=False, header=False)
    
    print(f"Cleaned CSV saved at: {destination_file}")
```

### match for dp03 across years for states

```python
import os
import pandas as pd

# Define directories
source_dir = r"H:\GY350\File Downloads"

# Iterate over years from 2010 to 2023
for year in range(2010, 2024):
    file_06 = os.path.join(source_dir, f"DP03_06_{year}.csv")
    file_32 = os.path.join(source_dir, f"DP03_32_{year}.csv")
    
    # Check if both files exist
    if not os.path.exists(file_06) or not os.path.exists(file_32):
        print(f"Skipping {year}: One or both files not found.")
        continue
    
    # Load CSVs
    df_06 = pd.read_csv(file_06, header=None)
    df_32 = pd.read_csv(file_32, header=None)
    
    # Check if the first row is identical
    if df_06.iloc[0].equals(df_32.iloc[0]):
        print(f"First row matches for {year}, updating DP03_32_{year}.csv")
        
        # Insert a new row after the first row, shifting all subsequent rows down
        new_row = df_06.iloc[1].copy()
        df_32 = pd.concat([df_32.iloc[:1], pd.DataFrame([new_row]), df_32.iloc[1:]], ignore_index=True)
        
        # Save the updated file
        df_32.to_csv(file_32, index=False, header=False)
        print(f"Updated DP03_32_{year}.csv successfully.")
    else:
        print(f"First row does not match for {year}, no changes made.")
```

### delete files from 2020 onwards as shapefiles change

```python
import os

# Define directories
directories = [
    r"H:\GY350\CSVs Cleaned",
    r"H:\GY350\File Downloads"
]

# Years to delete
years_to_delete = {"2020", "2021", "2022", "2023"}

# Iterate over directories
for directory in directories:
    if not os.path.exists(directory):
        print(f"Directory not found: {directory}")
        continue
    
    for filename in os.listdir(directory):
        if any(filename.endswith(f"_{year}.csv") for year in years_to_delete):
            file_path = os.path.join(directory, filename)
            os.remove(file_path)
            print(f"Deleted: {file_path}")
```

### check consistency of tracts

```python
import os
import pandas as pd

# Define directory
source_dir = r"H:\GY350\CSVs Cleaned"

# Function to check GEO_ID consistency across years and elaborate inconsistencies
def check_consistency(state):
    base_year = 2010
    base_file = os.path.join(source_dir, f"DP03_{state}_{base_year}.csv")
    
    if not os.path.exists(base_file):
        print(f"Base file for {state} in {base_year} not found.")
        return
    
    base_df = pd.read_csv(base_file)
    
    if "GEO_ID" not in base_df.columns:
        print(f"GEO_ID column not found in {base_file}")
        return
    
    base_geo_ids = set(base_df["GEO_ID"].dropna())
    
    inconsistencies_found = False
    
    for year in range(2011, 2020):
        file_path = os.path.join(source_dir, f"DP03_{state}_{year}.csv")
        
        if not os.path.exists(file_path):
            print(f"File missing: {file_path}")
            continue
        
        df = pd.read_csv(file_path)
        
        if "GEO_ID" not in df.columns:
            print(f"GEO_ID column missing in {file_path}")
            continue
        
        current_geo_ids = set(df["GEO_ID"].dropna())
        
        if base_geo_ids != current_geo_ids:
            inconsistencies_found = True
            print(f"Inconsistency found in GEO_ID for state {state} in {year}")
            missing_in_year = base_geo_ids - current_geo_ids
            new_in_year = current_geo_ids - base_geo_ids
            
            if missing_in_year:
                print(f"Missing GEO_IDs in {year}: {missing_in_year}")
            if new_in_year:
                print(f"New GEO_IDs in {year}: {new_in_year}")
    
    if not inconsistencies_found:
        print(f"GEO_ID values are consistent across 2010-2019 for state {state}")

# Check consistency for states 06 and 32
check_consistency("06")
check_consistency("32")
```
